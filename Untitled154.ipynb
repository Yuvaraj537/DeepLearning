```python
import os
import numpy as np
import pandas as pd
import cv2
from PIL import Image
from sklearn.model_selection import train_test_split

# Deep Learning (Tensorflow/Keras)
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Misc
import random
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully")
```

```python
# Updating dataset path
dataset_path = "train"

image_paths = []
labels = []

label_mapping = {
    "Closed_Eyes": 0,
    "Open_Eyes": 1    
}

# Walk through each category folder
for category in ["Closed_Eyes", "Open_Eyes"]:
    category_path = os.path.join(dataset_path, category)

    for file in os.listdir(category_path):
        if file.endswith(".jpg") or file.endswith(".png"):
            image_paths.append(os.path.join(category_path, file))
            labels.append(label_mapping[category])

# Create DataFrame
df = pd.DataFrame({
    "image_path": image_paths,
    "label": labels
})

# Shuffle
df = df.sample(frac=1).reset_index(drop=True)

# Display Info
print(f"Total images:{len(df)}")
print(f"Label distribution:\n{df['label'].value_counts()}")
df.head()
```

```python
# Plot sample Images
def show_samples(df, label, title, num_samples=5):
    sample_df = df[df['label'] == label].sample(num_samples)

    plt.figure(figsize=(15, 3))
    for idx, row in enumerate(sample_df.itertuples()):
        if not os.path.exists(row.image_path):
            print(f" Missing file: {row.image_path}")
            continue

        img = cv2.imread(row.image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #  lowercase cv2

        plt.subplot(1, num_samples, idx + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"{title} #{idx+1}")
        
    plt.tight_layout()
    plt.show()

# Call the function OUTSIDE
show_samples(df, label=1, title="Open Eye")
show_samples(df, label=0, title="Closed Eye")
```

```python
from torchvision import transforms

# Constants
IMAGE_SIZE = 64
BATCH_SIZE = 32

# Define transform
transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,)) # For RGB
])
```

```python
import torch
from torch.utils.data import Dataset
from PIL import Image
import os

class EyeDataset(Dataset):
    def __init__(self, df, transform=None, grayscale=True):
        self.df = df
        self.transform = transform
        self.grayscale = grayscale
        
    def __len__(self):
        return len(self.df)
        
    def __getitem__(self, idx):
        img_path = self.df.iloc[idx]['image_path']
        label = self.df.iloc[idx]['label']

        # Load Image using PIL
        mode = 'L' if self.grayscale else 'RGB'
        image = Image.open(img_path).convert(mode)

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)
```

```python
from torch.utils.data import Dataset, DataLoader
# Train -Val  Split
train_df, val_df = train_test_split(df, test_size=0.2,stratify=df['label'], random_state=12)

# Create datasets
train_dataset = EyeDataset(train_df, transform=transform)
val_dataset = EyeDataset(val_df, transform=transform)

# Create  dataloaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)


# Sample Check
print(f"Train size : {len(train_dataset)}, val size:{len(val_dataset)}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class DrowsinessCNN(nn.Module):
    def __init__(self):
        super(DrowsinessCNN, self).__init__()
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),  # 1 channel for grayscale
            nn.ReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)
        )

        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 8 * 8, 256),  # For 64x64 input
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 2)  # 2 classes: open/closed
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Initialize model, loss, optimizer
model = DrowsinessCNN().to(device) 
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Accuracy function
def calculate_accuracy(preds, labels):
    _, predicted = torch.max(preds, 1)
    return (predicted == labels).float().mean().item()

# Training function
def train(model, train_loader, val_loader, criterion, optimizer, epochs=10):
    for epoch in range(epochs):
        model.train()
        train_loss, train_acc = 0.0, 0.0 

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            acc = calculate_accuracy(outputs, labels)

            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            train_acc += acc

        train_loss /= len(train_loader)  
        train_acc /= len(train_loader)

        # Validation
        model.eval()
        val_loss, val_acc = 0.0, 0.0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)  
                outputs = model(images)
                loss = criterion(outputs, labels)
                acc = calculate_accuracy(outputs, labels)

                val_loss += loss.item()
                val_acc += acc

        val_loss /= len(val_loader)
        val_acc /= len(val_loader)

        print(f"Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")
```

```python
# start training
EPOCHS = 5
train(model, train_loader, val_loader, criterion, optimizer, epochs=EPOCHS)
 # save model
MODEL_PATH = "drowsiness_cnn.path"
torch.save(model.state_dict(), MODEL_PATH)
print(f"/n Model  Saved to {MODEL_PATH}")
```
